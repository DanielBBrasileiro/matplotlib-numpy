```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)

plt.plot(x,x, label = "Linear")
plt.plot(x, x**2, label = "Quadrado")

plt.title("Abordagem Pyplot")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")

plt.legend()

plt.show()
```
